Realice las transformaciones que considere pertinentes con el fin de preparar los datos para un modelo de clasificación de un paciente con desenlace en Hospitalización. 

```python
df.drop(['DIAS HOSPITALIZACION MQ','DIAS HOSPITALIZACI�N UPC'],axis=1,inplace=True) #variables redundantes ya que dan informacion despues de ser hospitalizado
# dependen de la hospitalizacion y no proporcionan pistas hacia el target
```

```python
#Se usara SelectKBest para reducir el ruido y el numero de variables que se alimentaran al modelo.
from sklearn.feature_selection import SelectKBest, chi2
skb=SelectKBest(chi2,k=5) # se usa k=5 ya que en practicas pasadas me ha resultado como el numero donde la mejora en el modelo se aplana
```

```python
X=df.drop('HOSPITALIZACION',axis=1)
y=df['HOSPITALIZACION']
```

```python
skb.fit(X,y)
X_new=skb.transform(X)
X_new
```

```python
X_new.shape
```

```python
for i, score in enumerate(skb.scores_): #scores 
    print(f'columna {i},score {score}') #mostramos los scores de cada columna
```

```python
indices = skb.get_support(indices=True)
print(indices)
selected_feature_names = X.columns[indices]
selected_feature_names #sacamos los nombres de las columnas
```

```python
X_new=pd.DataFrame(X_new,columns=selected_feature_names)
X_new['Hospitalizacion']=df['HOSPITALIZACION'] #agragamos la columna del target
```

```python
df['HOSPITALIZACION'].value_counts()
```

```python
y.value_counts()
```

```python
for column in X_new.columns:
    print(X_new[column].value_counts()) #cambio el numero de hospitalizaciones, redujeron las negativas y aumentaron las positivas por alguna razon
```

```python
#me da resultados raros con el selectKBest, asi que tomaremos las columnas que selecciono como importantes y a partir de eso creamos un nuevo df
df_reducido=df[['NUMERO DE DIAS POST BIOPSIA EN QUE SE PRESENTA LA COMPLICACI�N INFECCIOSA',
       'FIEBRE', 'ITU', 'TIPO DE CULTIVO', 'AGENTE AISLADO','HOSPITALIZACION']]
```
